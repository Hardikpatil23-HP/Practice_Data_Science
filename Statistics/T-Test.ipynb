```python
from sklearn.datasets import load_iris
import numpy as np
from scipy.stats import ttest_1samp

# Load the Iris dataset
iris = load_iris()
sepal_lengths = iris.data[:, 0]  # Sepal length feature

# Hypothetical population mean
population_mean = 6.0

# Perform one-sample t-test
t_stat, p_value = ttest_1samp(sepal_lengths, popmean=population_mean)

print(f"T-Statistic: {t_stat:.3f}")
print(f"P-Value: {p_value:.3f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample and population mean.")
```

```python
from sklearn.datasets import load_iris
from scipy.stats import ttest_ind

# Load the Iris dataset
iris = load_iris()
data = iris.data
target = iris.target

# Extract sepal lengths for Setosa (class 0) and Versicolor (class 1)
setosa_sepal_length = data[target == 0, 0]
versicolor_sepal_length = data[target == 1, 0]

# Perform independent two-sample t-test
t_stat, p_value = ttest_ind(setosa_sepal_length, versicolor_sepal_length)

print(f"T-Statistic: {t_stat:.3f}")
print(f"P-Value: {p_value:.3f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The means of sepal lengths are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the means.")
```

```python
from sklearn.datasets import load_iris
from scipy.stats import ttest_rel
import numpy as np

# Load the Iris dataset
iris = load_iris()
sepal_width = iris.data[:, 1]  # Sepal width feature

# Simulate "before" and "after" data by adding noise
before = sepal_width
after = sepal_width + np.random.normal(0, 0.2, len(sepal_width))  # Adding random noise

# Perform paired t-test
t_stat, p_value = ttest_rel(before, after)

print(f"T-Statistic: {t_stat:.3f}")
print(f"P-Value: {p_value:.3f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the two measurements.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the two measurements.")
```

```python
import numpy as np
import scipy.stats as stats

# Sample data (positively skewed)
data = np.random.exponential(scale=1, size=100)

# Check normality
print(stats.shapiro(data))  # Shapiro-Wilk test for normality

# Log transform the data
transformed_data = np.log(data)

# Check normality of transformed data
print(stats.shapiro(transformed_data))
```
